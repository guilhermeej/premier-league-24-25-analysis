```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_player_info = pd.read_csv('datasets/premier_player_info.csv')
df_player_stats = pd.read_csv('datasets/player_stats_2024_2025_season.csv')
```

```python
(df_player_info['player_name'] == df_player_stats['player_name']).value_counts()
```

```python
df_merged = pd.merge(df_player_stats, df_player_info, on='player_name', how='left')
```

```python
df_goals = df_merged[df_merged['Goals'] >= 1]
fig, ax = plt.subplots(figsize=(19, 5))
sns.histplot(data=df_goals, x="Nationality", kde=True)
plt.xticks(rotation=90)
plt.show()
```

```python
df_assists = df_merged[df_merged['Assists'] >= 1]
fig, ax = plt.subplots(figsize=(19, 5))
sns.histplot(data=df_assists, x="Nationality", kde=True)
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.scatterplot(data=df_merged, x="Goals", y="Minutes Played", hue="Preferred Foot")
ax.set_ylabel("Minutos Jogados",fontsize=12)
ax.set_xlabel("Gols")
ax.set_title("Minutos Jogados vs Gols")
ax.legend(title='Pé Preferido', labels=['Direito', 'Ambos', 'Esquerdo'])
plt.show()
```

```python
df_merged["Appearances"] = np.where(df_merged["Appearances"] == 0, df_merged["appearances_"], df_merged["Appearances"])
```

```python
df_merged.drop('appearances_', axis = 1, inplace=True)
```

```python
np.unique(df_merged['player_position'])
```

```python
mapa = {
    "Defender": "Defesa",
    "Midfielder": "Meio",
    "Forward": "Atacante",
    "Goalkeeper": "Goleiro"
}

df_merged["posicao_pt"] = df_merged["player_position"].map(mapa)

ig, ax = plt.subplots(figsize=(15, 5))
sns.scatterplot(data=df_merged, x="Goals", y="Appearances",hue="posicao_pt")
ax.set_ylabel("Jogos",fontsize=12)
ax.set_xlabel("Gols")
ax.set_title("Jogos vs Gols")
ax.legend(title='Posição')
plt.show()
```
